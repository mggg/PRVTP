```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def oneslate_pi(k, pi, n_voters=1, verbose=False):
  a_seats, b_seats = 0, 0
  a , b = pi, 1 - pi
  ratio = []
  while a_seats + b_seats < k:
    #print(a_seats, b_seats)
    if verbose: print(a,b, a/(a+b))
    ratio.append(a/(a+b))

    if a > b :
      a_seats += 1
      a_old = a
      a = (a_old - 1/(k+1)) * pi
      b =  b + (a_old - 1/(k+1))*(1 - pi)
    else:
      b_seats += 1
      b_old = b
      a = a + (b_old - 1/(k+1))*pi
      b = (b_old - 1/(k+1) ) * (1 - pi)

  if verbose: plt.plot(range(k), ratio)
  return a_seats

def oneslate(k):
  seats = [oneslate_pi(k, pi/101) for pi in range(0,101,1)]
  return seats
```

```python
def twoslate_pi(k, pi, a_share=1, n_voters=1, verbose=False):
  a_seats, b_seats = 0, 0
  a_from_a , b_from_a = a_share*pi, a_share*(1 - pi)
  a_from_b, b_from_b = (1-a_share)*(1 - pi), (1-a_share)*pi
  ratio = []
  while a_seats + b_seats < k:


    if a_from_a + a_from_b > b_from_a + b_from_b :
      a_seats += 1
      ratio = a_from_a / (a_from_a + a_from_b)
      a_from_a_old , a_from_b_old = a_from_a , a_from_b
      a_from_a = (a_from_a_old - ratio/(k+1)) * pi
      a_from_b = (a_from_b - (1-ratio)/(k+1)) * (1-pi)
      #a =  (a_from_a_old - ratio/(k+1)) * pi + (a_from_b - (1-ratio)/(k+1)) * (1-pi)
      b_from_a = b_from_a + (a_from_a_old - ratio/(k+1))*(1 - pi)
      b_from_b = b_from_b + (a_from_b_old - (1-ratio)/(k+1))*pi
      #b =  b + (a_from_a_old - ratio/(k+1))*(1 - pi) + (a_from_b - (1-ratio)/(k+1)) * pi
    else:
      b_seats += 1
      ratio = b_from_b / (b_from_b + b_from_a)
      b_from_a_old , b_from_b_old = b_from_a , b_from_b
      b_from_b = (b_from_b_old - ratio/(k+1))*pi
      b_from_a = (b_from_a_old - (1-ratio)/(k+1)) * (1 - pi)
      a_from_a = a_from_a + (b_from_a_old - (1-ratio)/(k+1)) * pi
      a_from_b = a_from_b + (b_from_b_old - ratio/(k+1))*(1-pi)
      #a = a + (b_from_a_old - (1-ratio)/(k+1))*pi + (b_from_b_old - ratio/(k+1))
      #b = (b_old - 1/(k+1) ) * (1 - pi)

  if verbose: plt.plot(range(k), ratio)
  return a_seats

def twoslate(k, a_share=1):
  seats = [twoslate_pi(k, pi/100, a_share) for pi in range(0,100,1)]
  return seats
```

```python
def twoslate_pi_uniform(k, pi, a_share=1, n_voters=1, num_cands=2, verbose=False):
  a_seats, b_seats = 0, 0
  a_cands, b_cands = k*num_cands,num_cands*k
  a_from_a , b_from_a = a_share*pi, a_share*(1 - pi)
  a_from_b, b_from_b = (1-a_share)*(1 - pi), (1-a_share)*pi
  ratio = []
  while a_seats + b_seats < k:


    if (a_from_a + a_from_b)/(a_cands - a_seats) > (b_from_a + b_from_b)/(b_cands - b_seats) :

      ratio = a_from_a / (a_from_a + a_from_b)
      a_from_a_old , a_from_b_old = a_from_a , a_from_b
      a_from_a = a_from_a_old *(a_cands - a_seats-1) /(a_cands - a_seats) + (a_from_a_old/(a_cands - a_seats) - ratio/(k+1)) * pi
      a_from_b = a_from_b_old *(a_cands - a_seats-1) /(a_cands - a_seats) + (a_from_b_old/(a_cands - a_seats) - (1-ratio)/(k+1)) * (1-pi)
      #a =  (a_from_a_old - ratio/(k+1)) * pi + (a_from_b - (1-ratio)/(k+1)) * (1-pi)
      b_from_a = b_from_a + (a_from_a_old/(a_cands - a_seats) - ratio/(k+1))*(1 - pi)
      b_from_b = b_from_b + (a_from_b_old/(a_cands - a_seats) - (1-ratio)/(k+1))*pi
      #b =  b + (a_from_a_old - ratio/(k+1))*(1 - pi) + (a_from_b - (1-ratio)/(k+1)) * pi
      a_seats += 1
    else:

      ratio = b_from_b / (b_from_b + b_from_a)
      b_from_a_old , b_from_b_old = b_from_a , b_from_b
      b_from_b = b_from_b_old* (b_cands - b_seats-1)/(b_cands - b_seats) + (b_from_b_old/(b_cands - b_seats) - ratio/(k+1))*pi
      b_from_a = b_from_a_old* (b_cands - b_seats-1)/(b_cands - b_seats) + (b_from_a_old/(b_cands - b_seats) - (1-ratio)/(k+1)) * (1 - pi)
      a_from_a = a_from_a + (b_from_a_old/(b_cands - b_seats) - (1-ratio)/(k+1)) * pi
      a_from_b = a_from_b + (b_from_b_old/(b_cands - b_seats) - ratio/(k+1))*(1-pi)
      #a = a + (b_from_a_old - (1-ratio)/(k+1))*pi + (b_from_b_old - ratio/(k+1))
      #b = (b_old - 1/(k+1) ) * (1 - pi)
      b_seats += 1
  if verbose: plt.plot(range(k), ratio)
  return a_seats

def twoslate_uniform(k, a_share=1, num_cands=2):
  seats = [twoslate_pi_uniform(k, pi/100, a_share, num_cands=num_cands) for pi in range(0,100,1)]
  return seats
```
